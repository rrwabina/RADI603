# Epidemiological Dynamics of the Monkeypox Virus using Ordinary Differential Equations



### **Section 1. Domain Problem** 

The emergence of infectious diseases with pandemic potential has served as a wake-up call to various governments across the globe. Througout human history, the biological threat of a pandemic disease has been devastating on an unprecedented scale, as evidenced by diseases such as COVID-19, influenza, and the bubonic plague. Because of these risks, the rapid spread of the Monkeypox disease this year has posed another global threat during COVID-19 pandemic. The Monkeypox, primarily caused by the monkeypox virus, is an infectious zoonotic disease which is closely related to smallpox but with milder infection. The World Health Organization (WHO) already declared the Monkeypox outbreak an international public health emergency, citing with a staggering increase of 30,000 cases worldwide. While the WHO has not declared the monkeypox oubreak a pandemic, several epidemiologists have highlighted the severity of monkeypox disease infection among the human population.

The transmission of monkeypox virus occurs when a person comes into contact with an infected animal or human. The virus is highly susceptible (i.e., able to contract the disease) to various animal species, particularly on non-human primates (e.g. monkeys). Animal-to-human transmission occurs primarily when infected monkeys transmit the monkeypox virus via bite, scratch, or even contaminated monkey meat preparation. Human-to-human transmission also occurs when infected humans pass the virus on to other people through direct contact with bodily fluids (i.e., sweat, semen, blood). Those who are susceptible can be exposed to the virus (i.e., have been infected but are not yet infectious) and will become infected (i.e., capable of transmitting the disease to other humans) at any time. Unfortunately, there is no known monkeypox vaccine in circulation, but smallpox vaccine may be used as a protection against Monkeypox with 85% effectivity. Nonetheless, while most infected people may recover from Monkeypox, the WHO highly recommends infected patients to take vaccine, especially if symptoms are severe.


#### 1.1 Derivatives in Disease Modeling

Controlling infectious diseases is a difficult task because the behavior of virus transmission among animal and human populations changes over time. Consider the following example to better understand the dynamics of virus transmission. Assume that the monkeypox virus has infected only one animal. This infected animal may scratch two people, exposing them to the virus, which may infect other people as well. Through direct contact, the infected humans may spread the Monkeypox virus to the other two people. Each person can then infect two more people, and so on. Because only one monkey is infected and humans can only infect two people, epidemiologists will be able to easily track virus transmission. However, how can epidemiologists track the spread of the Monkeypox virus if the source of infection is four infected monkeys? What if infected people spread the virus to four or six people? What if the virus evolves into a more potent mutation? In practice, these changes will be difficult to monitor because epidemiologists must account for them in order to explain the dynamics of virus transmission.  

Ordinary differential equations (ODEs) are the primary tools used to investigate changes in virus transmission. In disease modeling, these ODEs are known as epidemiological models that use compartments, such as susceptible $S$, infected $I$, recovered $R$, and etc., to explain the instantaneous change in the number of infected people over a fixed time $t$ - that is, a change in functions $S(t)$, $I(t)$, and $R(t)$. Because ODEs can examine a changing system (i.e., virus transmission) that includes the derivative of a function (e.g., change in number of infected humans), as well as the function (e.g., number of infected humans), these epidemiological models can be used to evaluate the long-term incidence of disease transmission among specific populations (e.g., number of infected humans). 

The key goal of epidemiological models is to calculate the long-term incidence of a disease. It refers to the number of people infected with a disease during a specific time period. It also looks into population behavior in terms of key epidemiological parameters (e.g., susceptible, infected, recovered) by examining when and how the epidemic will begin and end. Epidemiologists employ ODEs to track the number of infected people based on changes in other populations to solve long-term incidences. In the event of an epidemic or pandemic, this strategy will also assist local government units in implementing safety protocols and public health interventions. 

#### 1.2 Example Problem and Solution
In this example, we use the classical SIR model to analyze the spread of a disease. Suppose $S(t)$, $I(t)$, and $R(t)$ be the number of susceptible, infected, and recovered humans, respectively. The time $t$ is the number of days since the disease was discovered in an isolated town with a population of $N = 10,000$ people. 

Every SIR model requires parameters that characterize the change in the population of the compartments. In this example, we used three primary parameters, i.e., $\beta$, $\nu$, and $\gamma$. Suppose $\beta$ describes the transmission rate of the disease - that is, susceptible individual is in direct contact with an infected human. We also presume that $\gamma$ is the mean recovery rate and $\nu$ is the death rate of individuals caused by the disease. All parameters are assumed to be greater than zero. In addition, an individual is immuned and can no longer be susceptible when a person is recovered from the disease. For the sake of simplicity, this example ignores other parameters like the birth parameter $\Lambda$ and non-disease deaths $\nu$.

Our goal in this problem is to develop the differential equations based from the given assumptions to make predictions about the effects of the disease on the population. We apply the concepts of derivatives (rate of change) to formulate the differential equation to every compartment. The number of susceptible individuals $S_h(t)$ will decrease over time at a transmission rate $\beta$ proportional to the product of the number of susceptibles $S$ and number of infected $I$. Its differential equation is given as:
\begin{equation*}
    \frac{dS}{dt} = - \beta S(t)I(t) \\
    \tag{1}
\end{equation*}

The number of the infected people changes (i.e., $\frac{dI}{dt}$) increases with the transmission rate $\beta$ at which susceptible individuals become infectious. It is decreased by the death rate caused by the virus $\nu$, and the recovered individuals after infection due to recovery rate $\gamma$. Hence, the rate of change in the population of infected humans is:
\begin{equation*}
    \frac{dI}{dt} = \beta S(t)I(t) - (\nu + \gamma) I(t)
    \tag{2}
\end{equation*}

Finally, the number of recovered human population increases as a result of the transition from being clinically ill to the recovered compartments at the rate $\gamma$. Therefore, the rate of change in the population of recovered humans is:
\begin{equation*}
    \frac{dR}{dt} =  \gamma I(t)
    \tag{3} 
\end{equation*}


We can adopt the Euler's method to approximate the systems of equations above. Let $S(t)$, $I(t)$, and $R(t)$ be $S_t$, $I_t$, and $R_t$, respectively, to denote the number of days that have passed. We approximate $\frac{dS}{dt}$ by realizing that $\Delta S = S_{t+1} - S_t$. We also utilize the similar process for other compartments. This will result into the following:

\begin{equation*}
\begin{aligned}
    \Delta S        &= S_{t+1} - S_t \\
    - \beta S_tI_t    &= S_{t+1} - S_t \\
    S_{t+1}         &= S_t - \beta S_tI_t
    \tag{4}
\end{aligned}
\end{equation*}

\begin{equation*}
\begin{aligned}
    \Delta I        &= I_{t+1} - I_t \\
    \beta S_tI_t - (\nu + \gamma) I_t  &= I_{t+1} - I_t\\
    I_{t+1} &= I_t + \beta S_tI_t - (\nu + \gamma) I_t \\
    I_{t+1} &= I_t(1 + \beta S_t - (\nu + \gamma)) \\
    \tag{5}
\end{aligned}
\end{equation*}


\begin{equation*}
\begin{aligned}
    \Delta R        &= R_{t+1} - R_t \\
    \gamma I_t  &= R_{t+1} - R_t\\
    R_{t+1} &= R_t + \gamma I_t
    \tag{6}
\end{aligned}
\end{equation*}

Therefore, equations derived using Euler's method are expressed in Equation 7:
\begin{equation*}
\begin{aligned}
    S_{t+1}         &= S_t - \beta S_tI_t \\
    I_{t+1} &= I_t(1 + \beta S_t - \gamma -\nu)\\
    R_{t+1} &= R_t + \gamma I_t
    \tag{7}
\end{aligned}
\end{equation*}

**Suppose the transmission rate and death rate of the disease is $0.08$ and $0.0025$ per capita, respectively. The recovery rate of the infected individuals per capita is $0.0025$. What is $S(t)$, $I(t)$, and $R(t)$ in the first day of an epidemic outbreak if there are 25 confirmed cases during day 0 in a small town with $10,000$ population? What is the long-term incidence of the disease if $T = 1000$ days?**

Set the initial conditions at time $t = 0$ such that it assumes the following condition $\frac{S(0)}{N} + \frac{I(0)}{N} + \frac{R(0)}{N}= \frac{N}{N}$. Therefore, $\frac{I(0)}{N} = \frac{25}{10000} = 0.0025$ and $\frac{S(0)}{N} = 1 - I(0) = 0.9975$. The epidemiological parameters are $\beta = 0.08$, $\nu = 0.0025$, and $\gamma = 0.0025$.

\begin{equation*}
\begin{aligned}
    t &= 0 \\
    S_{1} &= N(S_0 - \beta S_0I_0) = (10000)(0.9975 - (0.08 * 0.9975 * 0.0025)) = (10000)(0.9973005) \approx 9973\\
    I_{1} &= N(I_0(1 + \beta S_0 - \nu - \gamma)) = (10000)((0.0025)(1+ (0.08)(0.9975) - 0.0025 - 0.0025)) = (10000)(0.002687) \approx 27\\
    R_{1} &= N(R_0 + \gamma I_0) = (10000)(0 + (0.005)(0.0025)) = (10000)(0.0000125) \approx 0 
    \\
    \\
    t &= 1 \\
    S_{2} &= N(S_1 - \beta S_1I_1) = (10000)(0.9973005 - (0.08 * 0.9973005 * 0.002687)) = (10000)(0.99708612028452) \approx 9971\\
    I_{2} &= N(I_1(1 + \beta S_1 - \nu - \gamma)) = (10000)((0.002687)(1+ (0.08)(0.9973005) - 0.0025 - 0.0025)) = (10000)(0.00288794471548) \approx 29\\
    R_{2} &= N(R_1 + \gamma I_1) = (10000)(0.0000125 + (0.005)(0.002687)) = (10000)(0.000025935) \approx 0 
    \tag{8}
\end{aligned}
\end{equation*}


Therefore, on the first day of the epidemic, there are $9,973$ susceptible individuals, $27$ infected cases, and $0$ recovered patients. On the second day, the number of susceptible individuals decreases by two cases since these two humans are infected with the virus. However, the number of recovered people remains zero. 

We created a sample Python code, along with its corresponding plot, to provide a faster computation of the disease's long-term incidence.

```python
import numpy as np
import matplotlib.pyplot as plt

T = 1000            # Number of days
N = 10000           # Population
S = np.zeros(T)
I = np.zeros(T)
R = np.zeros(T)

# Initial Conditions
I[0] = 0.0025       # 25 cases / 10,000 cases = 0.0025
S[0] = 1 - I[0]     # S[0] + I[0] = 1
R[0] = 0

# Transmission Parameters 
nu = 0.0025          # Death rate caused by the infection
beta = 0.08          # Transmission Rate
gamma = 0.0025       # Recovery rate (i.e., Infection to Recovered compartments)
time = np.array(range(0, T)) # Sequence of days

for i in range(T - 1):
    # Euler's method
    S[i + 1] = S[i] - (beta * S[i] * I[i])
    I[i + 1] = I[i] * (beta * S[i] - nu - gamma + 1)
    R[i + 1] = R[i] + gamma * I[i]

a = N

plt.figure(figsize = (8, 6))
plt.plot(S * a, 'b', label = 'Susceptible', lw = 3)
plt.plot(I * a, 'r', label = 'Infected', lw = 3)
plt.plot(R * a, 'y', label = 'Recovered', lw = 3)
plt.xlabel('Days', size = 15)
plt.ylabel('Number of cases', size = 15)
plt.grid(0)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.legend()
plt.title('Long-term incidence of a disease in Humans', size = 15)
```

### **Section 2. Formulation of the Mathematical Problem**

#### 2.1 Model Formulation

In this section, we formulate the epidemiological model of the monkeypox virus using the underlying presumptions from Section 1. We consider both animal-to-human and human-to-human transmission since the monkeypox infection involves monkey and human populations. Let's assume $S_h(t)$, $I_h(t)$, and $R_h(t)$ be the number of susceptible, infected, and recovered humans, respectively, where time $t$ is the number of days since the monkeypox epidemic was discovered. Since monkeypox infections have a significant latency period where infected individuals are not yet infectious themselves, the epidemiological model requires an **exposed** $E_h(t)$ compartment to describe this scenario. However, no intervention strategies will be included in this model (i.e., vaccinated compartments $V$) since existing vaccines are only created for smallpox infections. Therefore, this model only involves four components, having an SEIR model. 

The monkeypox virus can be transmitted to the susceptible human population from an infected monkey $I_m$ through a direct contact from $I_m$ to $S_h$. While it is feasible to formulate another SIR model for the monkey population, we just assume $I_m$ as a constant value since we focus on the monkeypox transmission among the human population. In addition, we adopt natural birth rates $\Lambda$ and natural death rate $\mu$ to provide a real-world simulation of the Monkeypox dynamics. 

Our goal in this problem is to develop the differential equations based from the given assumptions to make predictions about the effects of the disease on the population. We provide a schematic diagram of the SEIR model to describe the transmission pathway of the monkeypox virus among infected monkeys and the human population. 

<p align = 'center'>
<img src = 'figs\MainProblem.jpg' style = 'center'>

The number of susceptible humans $S_h(t)$ changes as a result of recruitment through human birth rate $\Lambda$ with respect to time $t$. However, $S_h(t)$ is decreased by $\theta$ due to a direct contact with an infected monkey $I_m$ and susceptible human $S_h$. The $S_h(t)$ also decreased due to natural death per capita rate $\mu$. Hence, the rate of change in the susceptible humans is described in the following expression:
$$
    \frac{dS_h}{dt} = \Lambda - \theta S_h(t)I_m - \mu S_h(t)  \tag{9}
$$

The number of exposed individuals $E_h(t)$ increases with the contact rate $\theta$ due to interaction of $S_h(t)$ and $I_m$ with respect to time $t$. The $E_h(t)$, however, can be decreased by the transmission rate (human-to-human) $\beta_h$ and natural death rate $\mu$. The death rate caused by the virus $\nu$ is not included in this equation since exposed individuals are not yet infectious and only experience mild symptoms. The rate of change in the exposed humans is described in the following expression:
$$
    \frac{dE_h}{dt} = \theta S_h(t)I_m - E_h(t)(\mu + \beta_h) \tag{10}
$$

The number of infected human population $I_h(t)$ changes as a result of the transition from being exposed $E_h(t)$ to highly infectious with a rate of $\beta_h$ with respect to time $t$. The $I_h(t)$ is decreased due to the proportion of individuals who either died from natural death, infection, or transitioned to being recovered. Hence, the rate of change in the infected compartment is described below:
$$
    \frac{dI_h}{dt} = \beta_h E_h(t) - (\mu + \nu + \gamma)  I_h(t) \tag{11}
$$

Finally, the number of recovered individuals increases due to transition from being infected $I_h(t)$ to being recovered with a recovery rate $\gamma$ with respect to time $t$. However, it may be decreased due to natural deaths of the recovered population. Therefore, the rate of change in the recovered compartment is expressed as:
$$
    \frac{dR_h}{dt} = \gamma I_h(t) -\mu R_h(t) \tag{12}
$$

Therefore, the epidemiological model of the monkeypox virus is defined in Equation 13. 
\begin{equation*}
\begin{aligned}
    \frac{dS_h}{dt} & = \Lambda - \theta S_h(t)I_m - \mu S_h(t)  \\
    \frac{dE_h}{dt} & = \theta S_h(t)I_m - E_h(t)(\mu_h + \beta_h) \\
    \frac{dI_h}{dt} & = \beta_h E_h(t) - (\mu + \nu + \gamma)  I_h(t)  \\
    \frac{dR_h}{dt} & = \gamma I_h(t) -\mu R_h(t)
    \tag{13}
\end{aligned}
\end{equation*}

Let $\beta_h = 0.016744$ while the recovery rate of the infected patients is $\gamma = 0.004$. Suppose there are 10 exposed individuals during $t = 10$. Evaluate the long-term incidence of the Monkeypox disease for $1000$ days using the formulated SEIR epidemiological model.

### III. Calculation

We utilize Euler's method to solve the differential equations of the monkeypox virus' epidemiological model. Suppose $S(t)$, $E(t)$, $I(t)$, and $R(t)$ be $S_t$, $E_t$, $I_t$, and $R_t$, respectively. Here, we want to calculate the dependent variables $S, E, I,$ and $R$. Therefore,

\begin{equation*}
\begin{aligned}
    \Delta S_{h_t}        &= S_{h_{t+1}} - S_{h_t} \\
    \Lambda - \theta S_{h_t} I_{m}    &= S_{h_{t+1}} - S_{h_t} \\
    S_{h_{t+1}}     &= S_{h_t} + \Lambda - \theta S_{h_t} I_{m}
    \tag{14}
\end{aligned}
\end{equation*}
<br>

\begin{equation*}
\begin{aligned}
    \Delta E_{h_t}    &= E_{h_{t+1}} - E_{h_t} \\
    \theta S_{h_t}I_m - E_{h_t}(\mu_h + \beta_h)  &= E_{h_{t+1}} - E_{h_t} \\
    E_{h_{t+1}}     &= E_{h_t} + \theta S_{h_t}I_m - E_{h_t}(\mu_h + \beta_h)
    \tag{15}
\end{aligned}
\end{equation*}
<br>

\begin{equation*}
\begin{aligned}
    \Delta I_{h_t}    &= I_{h_{t+1}} - I_{h_t} \\
    \beta_h E_{h_t} - (\mu + \nu + \gamma)  I_{h_t} &= I_{h_{t+1}} - I_{h_t} \\
    I_{h_{t+1}}     &= I_{h_t} + \beta_h E_{h_t} - (\mu + \nu + \gamma)  I_{h_t}
    \tag{16}
\end{aligned}
\end{equation*}
<br>

\begin{equation*}
\begin{aligned}
    \Delta R_{h_t}    &= R_{h_{t+1}} - R_{h_t} \\
    \gamma I_{h_t} -\mu R_{h_t} &= R_{h_{t+1}} - R_{h_t} \\
    R_{h_{t+1}}     &= R_{h_t} + \gamma I_{h_t} -\mu R_{h_t}
    \tag{17} 
\end{aligned}
\end{equation*}

Therefore, the derived equations from Euler's method is defined in the following expression:
\begin{equation*}
\begin{aligned}
    S_{h_{t+1}}     &= S_{h_t} + \Lambda - \theta S_{h_t} I_{m} \\
    E_{h_{t+1}}     &= E_{h_t} + \theta S_{h_t}I_m - E_{h_t}(\mu_h + \beta_h) \\
    I_{h_{t+1}}     &= I_{h_t} + \beta_h E_{h_t} - (\mu + \nu + \gamma)  I_{h_t} \\
    R_{h_{t+1}}     &= R_{h_t} + \gamma I_{h_t} -\mu R_{h_t}
    \tag{18}
\end{aligned}
\end{equation*}
Differential equations in (18) can also be expressed as a fraction of a compartment and the total population $N = 10,000$. Hence, 
\begin{equation*}
\begin{aligned}
    \frac{S_{h_{1}}}{N}     &= S_{h_0} + \Lambda - \theta S_{h_0} I_{m} \\
    \frac{E_{h_{1}}}{N}     &= E_{h_0} + \theta S_{h_0}I_m - E_{h_0}(\mu_h + \beta_h) \\
    \frac{I_{h_{1}}}{N}     &= I_{h_0} + \beta_h E_{h_0} - (\mu + \nu + \gamma)  I_{h_0} \\
    \frac{R_{h_{1}}}{N}     &= R_{h_0} + \gamma I_{h_0} -\mu R_{h_0}
    \tag{19}
\end{aligned}
\end{equation*}

Solve for $t = 1$
\begin{equation*}
\begin{aligned}
    \frac{S_{h_{t+1}}}{10000}     &= (S_{h_0} + \Lambda - \theta S_{h_t} I_{m}) = 10000(0.99 + 0.0015 - (0.052466)(0.99)(0.15)) \approx 9827\\
    \frac{E_{h_{t+1}}}{10000}     &= E_{h_0} + \theta S_{h_t}I_m - E_{h_0}(\mu_h + \beta_h) = 10000(0.01 + (0.052466)(0.99)(0.15) - 0.01(0.001 + 0.016744)) \approx 176\\
    \frac{I_{h_{t+1}}}{10000}     &= I_{h_0} + \beta_h E_{h_t} - (\mu + \nu + \gamma)  I_{h_0} = 10000(0 + (0.016744)(0.01) - (0.001 + 0.0000456621 + 0.004)(0)) \approx 2 \\
    \frac{R_{h_{t+1}}}{10000}     &= R_{h_0} + \gamma I_{h_0} -\mu R_{h_0} = 10000(0 + (0.004)(0) - (0.001)(0)) = 0
    \tag{20}
\end{aligned}
\end{equation*}

Therefore, at time $t = 1$, there have been $9,827$ susceptible, $176$ exposed, and $2$ infected individuals caused by the Monkeypox virus. Unfortunately, there have been no recoveries at day $1$. We created a Python code, as presented below, to calculate for the long-term incidence of the Monkeypox virus based on the presumptions in Section 1.  

```python
T = 1000
N = 10000

S = np.zeros(T)
E = np.zeros(T)
I = np.zeros(T)
R = np.zeros(T)

# Initial conditions
E[0] = 100 / N
I[0] = 0 / N
R[0] = 0
S[0] = 1 - (E[0] + I[0])

# Parameters
beta_h = 0.016744  # Transmission rate
lambda_h = 0.0015   # Birth (recruitment) rate of humans
mu = 0.001          # Natural death rate of humans
theta = 0.052466   # Contact rate between infected monkeys and suceptible humans
Im = 0.15           # Constant condition of infected monkey
nu = 0.0000456621   # Monkeypox-induced death rate of humans
gamma = 0.004       # Recovery rate of infected humans

for i in range(T - 1):
    # Euler's methods
    S[i + 1] = S[i] + lambda_h - S[i] * (theta * Im + mu)
    E[i + 1] = E[i] + (theta * Im * S[i]) - (mu + beta_h) * E[i]
    I[i + 1] = I[i] + beta_h * E[i] - (mu + nu + gamma) * I[i]
    R[i + 1] = R[i] + gamma * I[i] - mu * R[i]

a = N
plt.figure(figsize=(8, 7))
plt.plot(S * a, 'b', label='Susceptible', lw=3)
plt.plot(E * a, 'black', label='Exposed', lw=3)
plt.plot(I * a, 'r', label='Infected', lw=3)
plt.plot(R * a, 'y', label='Recovered', lw=3)
plt.xlabel('Days', size=15)
plt.ylabel('Number of cases', size=15)
plt.grid(0)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend()
plt.title('Long-term incidence of Monkeypox Virus in Human Population', size=15)
```
